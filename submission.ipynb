```python
## Creating the submission file
import pandas as pd
import numpy as np

df = pd.read_csv('artifacts/predicted_data.csv')
df.head()
```

```python
submission = df[['Timestamp', 'Water_Consumption_Prediction']].copy()
submission['Timestamp'] = pd.to_datetime(submission['Timestamp'], format="%Y-%m-%d %H:%M:%S")
submission['Timestamp'] = submission['Timestamp'].dt.strftime("%d/%m/%Y %H")
submission = submission.rename({'Water_Consumption_Prediction':'Water_Consumption'}, axis=1)
```

```python
submission['Water_Consumption'] = round(submission['Water_Consumption'],2)
submission = submission.set_index('Timestamp')
submission.to_csv('artifacts/submission.csv')
submission
```
